```python
import os
import pandas as pd
import shutil

folders = ["Training", "Testing", "Validation"]
for folder in folders:
    file_path = f'C:\\Users\\matus\\PycharmProjects\\VisionTransformer\\data\\B. RFMiD_Challenge_Dataset\\B. RFMiD_Challenge_Dataset\\2. Groundtruths\\a. RFMiD_{folder}_Labels.csv'
    df = pd.read_csv(file_path)
    
    # Filter rows where 'ARMD' column has value 1
    armd_value_1_df = df[df['ARMD'] == 1]
    
    source_folder = f'C:\\Users\\matus\\PycharmProjects\VisionTransformer\data\B. RFMiD_Challenge_Dataset\B. RFMiD_Challenge_Dataset\1. Original Images'
    destination_folder = 'C:\\Users\\matus\\PycharmProjects\\VisionTransformer\\data\\binary\\Disease'
    
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)
    
    for index, row in armd_value_1_df.iterrows():
        # Extract the image name from the 'ID' column
        image_name = str(row['ID'])+ '.png'
    
        # Construct the full path of the image
        image_path = os.path.join(source_folder, image_name)
    
        # Check if the image exists
        if os.path.exists(image_path):
            # Get the base filename without extension
            filename, file_extension = os.path.splitext(image_name)
            
            # Check if the filename already exists in the destination folder
            count = 1
            new_filename = filename
            while os.path.exists(os.path.join(destination_folder, new_filename + file_extension)):
                new_filename = f"{filename}_{count}"
                count += 1
            
            # Construct the full path for the destination with the new filename
            destination_path = os.path.join(destination_folder, new_filename + file_extension)
            
            # Move the image to the destination folder
            shutil.move(image_path, destination_path)
            print(f"Moved {image_name} to {destination_path}")
        else:
            print(f"Image {image_name} not found in {source_folder}")
```
